```python
import keras 
import tensorflow
```

```python
from tensorflow.keras.preprocessing.text import Tokenizer
```

```python
tk = Tokenizer()
```

```python
tk.fit_on_texts(corpus)
```

```python
len(tk.word_index)
```

```python
len(tk.index_word)
```

```python
data = tk.texts_to_sequences(corpus)
```

```python
print(data)
```

```python
X = []
y = []

for li in data:
    for ind in range(1,len(li)):
        X.append(li[:ind])
        y.append(li[ind])
        
    
        
    
```

```python
import pandas as pd
```

```python
print(data)
```

```python
pd.DataFrame({'X':X,'y':y})
```

```python
len(X),len(y)
```

```python
from keras.utils import pad_sequences,to_categorical
```

```python
feature_variable = pad_sequences(X,padding='post')

```

```python
feature_variable.shape
```

```python
class_label = to_categorical(y,num_classes=36+1)
```

```python
class_label.shape
```

```python
from keras.models import Sequential
from keras.layers import Dense,Embedding,SimpleRNN,InputLayer
```

```python
model = Sequential()
model.add(InputLayer(shape=(23,)))
model.add(Embedding(37,2))
model.add(SimpleRNN(10,return_state=False))
model.add(Dense(37,activation='softmax'))

```

```python
model.summary()
```

```python
model.compile( optimizer='rmsprop',
    loss='categorical_crossentropy',
    metrics=['accuracy'])
```

```python
model.fit(feature_variable,class_label,epochs=500)
```

```python
import numpy as np
```

```python
X = ["Employed"]
np.argmax(model.predict(pad_sequences(tk.texts_to_sequences(X),maxlen=23)))
```

```python
X = "Employed"

for y in range(10):
    word = tk.index_word[np.argmax(model.predict(pad_sequences(tk.texts_to_sequences([X]),maxlen=23)))]
    X = X+" "+word
    print(X)
    
```
